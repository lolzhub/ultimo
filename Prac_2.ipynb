**List of Columns with Null Values:**

```python
cols_with_na =[]
for col in Academic_Performance.columns:
  if Academic_Performance[col].isnull().any():
    cols_with_na.append(col)

cols_with_na
```

### **Dealing with Missing values:**

**1. Eliminating null values using dropna()**

```python
Academic_Performance.dropna() #Dropping ROWS with atleast 1 NULL  value
```

```python
Academic_Performance
```

**2. using replace() with nan**

```python
Academic_Performance_replace=Academic_Performance.replace(np.nan, value=10)
Academic_Performance_replace
```

**3. Using only Fillna()**

```python
Academic_Performance_fillna1=Academic_Performance.fillna(10)
Academic_Performance_fillna1
```

**4. Using Fillna() with mean,median,min,max of that column**

```python
Academic_Performance_fillna2=Academic_Performance
```

```python
Academic_Performance_fillna2['AI']=Academic_Performance_fillna2['AI'].fillna(Academic_Performance_fillna2['AI'].mean())
Academic_Performance_fillna2['CC']=Academic_Performance_fillna2['CC'].fillna(Academic_Performance_fillna2['CC'].median())
Academic_Performance_fillna2['DSBDA']=Academic_Performance_fillna2['DSBDA'].fillna(Academic_Performance_fillna2['DSBDA'].min())
Academic_Performance_fillna2['WT']=Academic_Performance_fillna2['WT'].fillna(Academic_Performance_fillna2['WT'].max())
```

```python
Academic_Performance_fillna2
```

**1. Using Inter Quantile Range(IQR)**

```python
Q1 = Academic_Performance['Percentage'].quantile(0.25)
Q3 = Academic_Performance['Percentage'].quantile(0.75)
IQR = Q3-Q1

Lower_Limit = Q1 - 1.5 * IQR
Upper_Limit = Q3 + 1.5 * IQR
print(f'Q1 = {Q1}, Q3 = {Q3}, IQR = {IQR}, Lower Limit = {Lower_Limit}, Upper Limit = {Upper_Limit}')
```

```python
Academic_Performance[(Academic_Performance['Percentage'] < Lower_Limit) | (Academic_Performance['Percentage'] > Upper_Limit)]
```

```python
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (9, 6)
df_list = ['AI','CC','Total','Percentage']
fig, axes = plt.subplots(2, 2)
fig.set_dpi(120)

count=0
for r in range(2):
  for c in range(2):
    _= Academic_Performance[df_list[count]].plot(kind = 'box', ax=axes[r,c])
    count+=1
```
